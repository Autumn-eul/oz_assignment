```python
class Shop:
    def __init__(self):
        stock = {"팥붕어빵" : 10,
        "슈크림붕어빵" : 8,
        "초코붕어빵" : 5}

        price = {
            "팥붕어빵" : 1000,
            "슈크림붕어빵" : 1200,
            "초코붕어빵" : 1500,
        }
        
        sales = {
            "팥붕어빵" : 0,
            "슈크림붕어빵" : 0,
            "초코붕어빵" : 0,
        }
        
    def order_bread(self): # 주문을 받는 함수
        while True:
                bread_type = input("붕어빵 종류를 입력해주세요(팥붕어빵, 슈크림붕어빵, 초코붕어빵 중 하나를 선택해주세요) 뒤로 돌아가길 원하시면 돌아가기를 입력해주세요 :")
                
                if bread_type == "돌아가기": 
                    break
                
                if self.stock[bread_type] <= 0:
                    print(f'현재 {bread_type}의 재고가 부족하여 주문이 어렵습니다. 다른 붕어빵을 골라주세요.')
                    continue
                    
                bread_count = int(input("주문할 붕어빵 개수를 입력해주세요 : ")) #str "3", '3' int() 함수 => 특정 기능이 작동되도록 모아놓은 코드다 => 자료형을 숫자로 형태의 자료로 만드는 기능의 함수
                    
                if self.stock[bread_type] >= bread_count:
                   self.stock[bread_type] -= bread_count # stock["초코붕어빵"] = 5-3
                   self.sales[bread_type] += bread_count
                   print(f"{bread_type} {bread_count}개를 판매했습니다.")
                    
                else:
                    answer = input(f'죄송합니다. 현재 {bread_type}의 재고가 {self.stock[bread_type]}개 입니다. {self.stock[bread_type]}개만 구매하시겠습니까? (네/아니오)')
                    if answer == "네":
                        print(f'{bread_type} {self.stock[bread_type]}개가 판매 완료 되었습니다.') # 구매한 종류와 수량 출력
                        self.stock[bread_type] -= bread_count # 재고 감소
                        self.sales[bread_type] += bread_count # 판매량 업데이트
                    else:
                        print(f'현재 {bread_type}의 재고가 부족하여 주문이 어렵습니다. 다른 붕어빵을 골라주세요.')
                        
    def manage_stock(self): # 재고를 추가하는 함수
        while True:
                bread_type = input("추가할 붕어빵 종류를 입력해주세요(팥붕어빵, 슈크림붕어빵, 초코붕어빵 중 하나를 선택해주세요) 뒤로 돌아가길 원하시면 돌아가기를 입력해주세요 :")
                if bread_type == "돌아가기": # bread_type에는 초코붕어빵 
                    break
                bread_count = int(input("추가할 붕어빵 개수를 입력하세요: "))
                self.stock[bread_type] += bread_count #stock["초코붕어빵"] 5 + bread_count 10 => 15 => stock["초코붕어빵"] = 15 덮어씌우기
                print(f'{bread_type}의 재고가 {bread_count}개 추가되었습니다.\n현재 재고\n팥붕어빵 : {self.stock["팥붕어빵"]}개\n슈크림붕어빵 : {self.stock["슈크림붕어빵"]}개\n초코붕어빵 : {self.stock["초코붕어빵"]}개')
            
    def total_sale(self):
        total_sale = sum(sales[bread] * price[bread] for bread in sales)
        print(f'총 매출액은 {total_sale()}원 입니다.\n현재 재고\n팥붕어빵 : {self.stock["팥붕어빵"]}개\n슈크림붕어빵 : {self.stock["슈크림붕어빵"]}개\n초코붕어빵 : {self.stock["초코붕어빵"]}개')

    def main(self):
        while True:
            mode = input('원하는 모드를 선택해주세요(주문, 관리자, 종료):')
            
            if mode == "종료":
                print("프로그램이 종료 되었습니다.")
                if sales[bread_type] > 0:
                    self.total_sale()
                else:
                    break
            
            if mode == "주문":
                self.order_bread()
                
            if mode == "관리자":
                self.manage_stock()
                
                
bbshop = Shop()
bbshop.main()
```
